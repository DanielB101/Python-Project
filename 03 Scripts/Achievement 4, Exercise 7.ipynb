#### Price range conditions

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# What is the high price in my subset?
df['prices'].max()
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

#### Loops day/orders

```python
# Which day has the most orders?
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Define busiest, least busy, regularly busy
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append("Least Busy")
    else:
        result.append("Regularly Busy")
```

```python
result
```

```python
df_ords_prods_merge['Busiest Day'] = result
```

```python
df_ords_prods_merge['Busiest Day'].value_counts(dropna = False)
```

# Step 2

```python
# Define busiest days
result_days = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value in [0,1]:
        result_days.append("Busiest Days")
    elif value in[4,3]:
        result_days.append("Least Busy Days")
    else:
        result_days.append("Regularly Busy Days")
```

```python
# Combine it with df_ords_prods_merge df
df_ords_prods_merge['Busiest Days'] = result_days
```

```python
df_ords_prods_merge['Busiest Days'].value_counts()
```

# Step 4

```python
# Find busiest hours
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Define busiest hours/most orders; top 4 = busiest hours, bottom 4 = least busy hours
result_hours = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in [10,11,14,15]:
        result_hours.append("Busiest Hours")
    elif value in [3,4,2,5]:
        result_hours.append("Least Busy Hours")
    else:
        result_hours.append("Regularly Busy Hours")
```

# Step 5

```python
# Combine it with df_ords_prods_merge df
df_ords_prods_merge['Busiest Hours'] = result_hours
```
