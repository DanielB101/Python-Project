# Step 6 

```python
# Aggregation; creating a spending flag to identify High spenders (over 10)
ords_prods_merge['mean_order_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Deriving Columns; Spending Flags
ords_prods_merge.loc[ords_prods_merge['mean_order_price'] >= 10, 'spending_flag'] = 'Big spender'
ords_prods_merge.loc[ords_prods_merge['mean_order_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'spending_flag', 'mean_order_price']].head(100)
```

# Step 7

```python
# Aggregation; creating a spending flag to identify Non-Frequent, Frequent and Regular Customers
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Deriving Columns; Frequency Flags
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] <= 20) & (ords_prods_merge['median_prior_order'] > 10), 'frequency_flag'] = 'Regular-Frequency Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'spending_flag', 'mean_order_price', 'frequency_flag', 'median_prior_order']].head(100)
```
